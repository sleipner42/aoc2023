```python
import re
from collections import Counter
from functools import cmp_to_key
```

```python
with open("data.txt") as file:
    data = file.read()
data = data.splitlines()

cards = {
    "A":12, 
    "K":11, 
    "Q":10, 
    "J":9, 
    "T":8, 
    "9":7, 
    "8":6, 
    "7":5, 
    "6":4, 
    "5":3, 
    "4":2, 
    "3":1, 
    "2" :0,
}

def check_hand(hand):

    counter = Counter(hand)

    if set(counter.values()) == {5}:
        return 700
    elif set(counter.values()) == {4,1}:
        return 600
    elif set(counter.values()) == {2,3}:
        return 500
    elif set(counter.values()) == {3,1,1}:
        return 400
    elif sorted(counter.values()) == [1,2,2]:
        return 300
    elif set(counter.values()) == {2,1,1,1}:
        return 200
    elif set(counter.values()) == {1,1,1,1,1}:
        return 100

def best_hand(hand):
    counter = Counter(hand)
    c_set = set(counter.values())

    if (
        c_set == {5}
        or max([val + counter["J"] for key, val in counter.items() if key != "J"]) == 5
    ):
        return 700
    elif (
        c_set == {4, 1}
        or max([val + counter["J"] for key, val in counter.items() if key != "J"]) == 4
    ):
        return 600
    elif c_set == {2, 3} or (sorted(counter.values()) == [1,2,2] and counter["J"] == 1):
        return 500
    elif (
        c_set == {3, 1, 1}
        or max([val + counter["J"] for key, val in counter.items() if key != "J"]) == 3
    ):
        return 400
    elif sorted(counter.values()) == [1, 2, 2]:
        return 300
    elif c_set == {2, 1, 1, 1} or max([val + counter["J"] for key, val in counter.items() if key != "J"]) == 2:
        return 200
    elif c_set == {1, 1, 1, 1, 1}:
        return 100

def cmp_items(a, b):

    a_hand = check_hand(a.split()[0])
    b_hand = check_hand(b.split()[0])

    if a_hand > b_hand:
        return 1
    elif a_hand == b_hand:
        for n in range(0,5):
            if cards[a[n]] > cards[b[n]]:
                return 1
            elif cards[a[n]] < cards[b[n]]:
                return -1
        
        print("problem")

    elif a_hand < b_hand:
        return -1
    
print(sum([(n+1) * int(val.split()[-1]) for n, val in enumerate(sorted(data, key=cmp_to_key(cmp_items)))]))

cards['J'] = -1
check_hand = best_hand

print(sum([(n+1) * int(val.split()[-1]) for n, val in enumerate(sorted(data, key=cmp_to_key(cmp_items)))]))
```
