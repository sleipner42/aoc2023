```python
import re
import numpy as np

with open("data.txt") as file:
    data = file.read()
data = data.splitlines()
```

```python
def extrapolate(seq, frwd = True):

    seq = np.array(list(map(int,seq.split())))
    derivative = seq
    derivatives = [seq]

    while ((derivative := np.diff(derivative)) != 0).any():
        derivatives.append(derivative)

    extrapolated_nums = [0]

    if frwd:
        for n in derivatives[::-1]:
            exp_num = extrapolated_nums[-1] + n[-1]
            extrapolated_nums.append(exp_num)
    else:
        for n in derivatives[::-1]:
            exp_num = -extrapolated_nums[-1] + n[0]
            extrapolated_nums.append(exp_num)
            print(exp_num, n)

    return extrapolated_nums[-1]

sum([extrapolate(seq) for seq in data])
```
