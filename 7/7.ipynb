```python
import re
from collections import Counter
from functools import cmp_to_key
```

```python
with open("data.txt") as file:
    data = file.read()
data = data.splitlines()

cards = {
    "A":12, 
    "K":11, 
    "Q":10, 
    "J":9, 
    "T":8, 
    "9":7, 
    "8":6, 
    "7":5, 
    "6":4, 
    "5":3, 
    "4":2, 
    "3":1, 
    "2" :0,
}

star = 1

def cmp_items(a, b):

    a_cards = a.split()[0]
    b_cards = b.split()[0]

    if star == 1:
        a_hand = sorted(list(Counter(a_cards).values()))[::-1]
        b_hand = sorted(list(Counter(b_cards).values()))[::-1]
    else:

        if a == 'KKKKJ 18':
            print("hej")

        a_common = Counter(a_cards).most_common()[0][0]
        b_common = Counter(b_cards).most_common()[0][0]

        a_hand = sorted(list(Counter(a_cards.replace("J", a_common)).values()))[::-1]
        b_hand = sorted(list(Counter(b_cards.replace("J", b_common)).values()))[::-1]

    if a_hand > b_hand:
        return 1
    elif a_hand == b_hand:
        for n in range(0,5):
            if cards[a[n]] > cards[b[n]]:
                return 1
            elif cards[a[n]] < cards[b[n]]:
                return -1
        
        print("problem")

    elif a_hand < b_hand:
        return -1
    
print(sum([(n+1) * int(val.split()[-1]) for n, val in enumerate(sorted(data, key=cmp_to_key(cmp_items)))]))

cards['J'] = -1
star = 2

print(sum([(n+1) * int(val.split()[-1]) for n, val in enumerate(sorted(data, key=cmp_to_key(cmp_items)))]))
```
